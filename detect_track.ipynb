```python
# ==========================================================
# TASK 4: Object Detection & Tracking using YOLOv8 + ByteTrack
# Google Colab (Single Cell)
# ==========================================================

# Install dependencies
!pip -q install ultralytics opencv-python

import os
import glob
from ultralytics import YOLO
from google.colab import files

print("="*60)
print("UPLOAD YOUR VIDEO (.mp4)")
print("="*60)

uploaded = files.upload()

video_path = list(uploaded.keys())[0]
print(f"\nUploaded Video: {video_path}")

# Load YOLOv8 Nano Model
model = YOLO("yolov8n.pt")

print("\nStarting Object Detection + Tracking...\n")

# Perform detection + tracking
model.track(
    source=video_path,
    tracker="bytetrack.yaml",
    save=True,
    persist=True,
    conf=0.4,
    project="runs",
    name="tracking",
    exist_ok=True,
    show=False
)

print("\nProcessing Finished!\n")

# Search for output video
video_files = []

video_files.extend(glob.glob("runs/**/*.mp4", recursive=True))
video_files.extend(glob.glob("runs/**/*.avi", recursive=True))

# Remove original uploaded video if found
video_files = [
    f for f in video_files
    if os.path.basename(f) != os.path.basename(video_path)
]

if video_files:
    # Pick the most recently modified output
    output_video = max(video_files, key=os.path.getmtime)

    print("="*60)
    print("OUTPUT VIDEO FOUND")
    print(output_video)
    print("="*60)

    files.download(output_video)

else:
    print("="*60)
    print("No output video found.")
    print("Available files:")
    print("="*60)

    for root, dirs, files_list in os.walk("runs"):
        for file in files_list:
            print(os.path.join(root, file))
```
